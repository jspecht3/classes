## Question 1

```python
# part 3
def ea(r):
    return -1.527 / r

def er(r):
    return (6.121e-6) / r**8

req = (4.8968e-5 / 1.527)**(1/7)
eeq = ea(req) + er(req)

print("Eq Radius [nm]:", req)
print("Eq Energy [eV]:", eeq)
```

```python
# part 4
rs = np.linspace(0.16, 1, 1000)

eas = ea(rs)
ers = er(rs)
ens = eas + ers

plt.plot(rs, eas, label='Attractive Energy')
plt.plot(rs, ers, label='Repulsive Energy')
plt.plot(rs, ens, label='Net Energy')

plt.ylabel('Energy [eV]')
plt.xlabel('Distance [nm]')

plt.axvline(req, label='Equilibrium Distance', ls=(0,(5,3)), c='r', lw=1)
plt.axhline(eeq, label='Equilibrium Energy', ls=(0,(5,3)), c='k', lw=1)

plt.legend()
plt.savefig('hw1-q1dot4.png', dpi=600)
```
