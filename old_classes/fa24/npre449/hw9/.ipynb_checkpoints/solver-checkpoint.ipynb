```python
import numpy as np
import numpy.linalg as la
import scipy as sci
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
from pyXSteam.XSteam import XSteam as xs
```

```python
class InitialConditions:
    def __init__(self, reactor_type):
        """Defines the initial conditions for the reactor.
        
        Parameters
        ----------
        reactor_type: [1,2]
            defines the reactor type, 1 corresponds to a PWR and 2 corresponds to a BWR
        """
        
        if reactor_type == 1:
            self.__dict__ = {
                "reactor": "PWR",
                
                "h": 4, # m
                "d_rod": 0.95 / 100, # cm / 100 = m
                "pitch": 1.26 / 100, # cm / 100 = m
                "d_fuel": 0.82 / 100, # cm / 100 = m
                "t_gap": 0.006 / 100, # cm / 100 = m
                
                "k_gap": 0.25, # W / mK
                "k_fuel": 3.6, # W / mK
                "k_clad": 21.5, # W / mk
                
        
                "g": 4000, # kg / m^2 s
                "qp_0": 430 * 100, # W / cm * 100 = W / m
                "p_z0": 15, # MPa
                "tf_z0": 277 + 273, # *C => K
            }

        if reactor_type == 2:
            self.__dict__ = {
                "reactor": "BWR",
                
                "h": 4.1, # m
                "d_rod": 1.227 / 100, # cm / 100 = m
                "pitch": 1.62 / 100, # cm / 100 = m
                "d_fuel": 1.04 / 100, # cm / 100 = m
                "t_gap": 0.010 / 100, # cm / 100 = m
                
                "k_gap": 0.25, # W / mK
                "k_fuel": 3.6, # W / mK
                "k_clad": 21.5, # W / mk
        
                "g": 2350, # kg / m^2 s
                "qp_0": 605 * 100, # W / cm * 100 = W / m
                "p_z0": 7.5, # MPa
                "tf_z0": 272 + 273, # *C => K
            }
        
        if reactor_type not in [1,2]:
            raise AttributeError("----- !!! Unsupported Reactor Type !!! -----\n" +
                            "The supported reactor types are:\n" +
                            "For a PWR, reactor_type == 1\n" +
                            "For a BWR, reactor_type == 2")
```

```python
class EnthalpyFunctions:
    def __init__(self):
        """You shouldn't be here"""

        # enthalphy dictionary
        enthalpies = {
            "hg": [],
            "hf": [],
            "hfg": [],
            "pressures": np.linspace(0.001,22.06,10000)
        }

        # steam lookup table
        steamtable = xs(xs.UNIT_SYSTEM_BARE) # m/kg/s/K/MPa/W

        # obtaining the enthalpies as a function of pressure for liquid, gas, and 2-phase
        for pressure in enthalpies["pressures"]:
            hg = steamtable.hV_p(pressure)
            hf = steamtable.hL_p(pressure)
            hfg = hg - hf
        
            enthalpies["hg"].append(hg)
            enthalpies["hf"].append(hf)
            enthalpies["hfg"].append(hfg)

        # getting polynomial fits for the enthalpy functions
        poly_deg = 25
        self.steamtable = steamtable

        ## h(P)
        hg_vs_p = np.polynomial.Polynomial.fit(
            enthalpies["pressures"], enthalpies["hg"], deg=poly_deg)
        hf_vs_p = np.polynomial.Polynomial.fit(
            enthalpies["pressures"], enthalpies["hf"], deg=poly_deg)
        hfg_vs_p = np.polynomial.Polynomial.fit(
            enthalpies["pressures"], enthalpies["hfg"], deg=poly_deg)

        ## dh/dP
        dhg_dp = hg_vs_p.deriv()
        dhf_dp = hf_vs_p.deriv()
        dhfg_dp = hfg_vs_p.deriv()

        # initializing the dict
        self.__dict__ = {
            "hg": hg_vs_p, 
            "hf": hf_vs_p,
            "hfg": hfg_vs_p,

            "dhg": dhg_dp,
            "dhf": dhf_dp,
            "dhfg": dhg_dp,
        }

    def plot(self):
        x = np.linspace(0.001, 22.06, 10000)

        hg = self.hg(x)
        hf = self.hf(x)
        hfg = self.hfg(x)

        dhg = self.dhg(x)
        dhf = self.dhf(x)
        dhfg = self.dhfg(x)

        for enthalpy, title in zip([hg, dhg, hf, dhf, hfg, dhfg],
                                   ["h_g", "dh_g/dP", "h_f", "dh_f/dP", "h_fg", "dh_fg/dP"]):
            plt.plot(x, enthalpy)
            plt.xlabel("Pressure [MPa]")
            plt.ylabel(title)
            plt.grid('both')
            plt.savefig(f"plots/{title.replace('/', '')}", dpi=600)
            plt.show()
```

```python
class Solver:
    def __init__(self, reactor_type, mesh_elements = 101):
        self.rector_type = reactor_type

        self.ics = InitialConditions(reactor_type)
        self.enthalpies = EnthalpyFunctions()
        self.st = xs(xs.UNIT_SYSTEM_BARE) # m/kg/s/K/MPa/W
        
        ics = self.ics
        st = self.st

        self.xi = np.pi * ics.d_rod
        self.a_fluid = ics.pitch**2 - np.pi/4 * ics.d_rod**2
        self.d_h = 4 * self.a_fluid / self.xi

        self.rho0 = st.rhoL_p(ics.p_z0)
        
        self.mu_f = st.my_pt(ics.p_z0, ics.tf_z0)# * self.rho0 * 1e-3 # mPa*s / 1000 = Pa*s
        self.k_fluid = st.tc_pt(ics.p_z0, ics.tf_z0)
        self.cp = st.Cp_pt(ics.p_z0, ics.tf_z0)
        self.mw = 18
        
        self.re = ics.g * self.d_h / self.mu_f #6 * self.d_h / self.mu_f
        self.pr = self.cp * self.mu_f / self.k_fluid * 1000
        self.nu = 0.023 * self.re**(0.8) * self.pr**(0.4)
        
        self.h_f0 = self.nu / self.d_h * self.k_fluid
        self.ff = 0.316 * self.re**(-1/4)

        self.pc = 22.06 # MPa

        self.zs = np.linspace(0, self.ics.h, mesh_elements)

    def solve_fluid(self, show):
        # renaming
        ics = self.ics
        enthalpies = self.enthalpies
        st = self.st
        
        # initial calcs
        h = st.h_pt(ics.p_z0, ics.tf_z0)
        hf = st.hL_p(ics.p_z0)
        hg = st.hV_p(ics.p_z0)

        tf0 = self.ics.tf_z0
        tsat0 = st.tsat_p(self.ics.p_z0)

        p0 = self.ics.p_z0
        chi_e0 = self.cp * (tf0 - tsat0) / enthalpies.hfg(p0)
        
        # lists
        temp_fluid = [tf0]
        temp_sat = [tsat0]
        
        pressure = [p0]
        chi_e = [chi_e0]

        # functions
        self.heat_flux = lambda z : 1e-3 * ics.qp_0 / self.xi * np.sin(np.pi * z / ics.h) # W / kW
        self.vol_heat_gen = lambda z: 4 * ics.qp_0 / np.pi / ics.d_fuel**2 * np.sin(np.pi * z / ics.h) / 1000 # W / kW
        
        if show: print(f"z: {self.zs[0]} \ttf: {round(temp_fluid[0],2)}        \tp: {round(pressure[0],3)}       \tchi_e: {round(chi_e[0],3)}\tt_sat: {round(temp_sat[0],3)}")
        #z: 0.4 	tf: 551.62069 	p: 14.9797266593 	chi_e: -0.32 	t_sat: 615.2
        for i in range(1, len(self.zs)):
            # step vars
            z = self.zs[i]
            dz = z - self.zs[i-1]
            
            chi_e0 = chi_e[i - 1]
            tf0 = temp_fluid[i - 1]
            p0 = pressure[i - 1]

            # calcs
            qpp = self.heat_flux(z)

            # vapor quality
            if chi_e0 <= 0:
                chi = 0
                dchi = 0
            if chi_e0 > 0 and chi_e0 <= 1:
                chi = chi_e0
                chi_prev = chi_e[i - 1]
                if chi_prev < 0: chi_prev = 0
                dchi = chi - chi_prev
            if chi_e0 > 1:
                print("WARNING!!! DRYOUT HAS OCCURED") # yeay yeayesye ayea


            # densities
            nu_f = st.vL_p(p0) # maybe an issue here
            nu_g = st.vV_p(p0)
            nu_fg = nu_g - nu_f

            nu_m = nu_f + chi * nu_fg

            rho_l = st.rhoL_p(pressure[0])
            rho_g = st.rhoV_p(pressure[0])
            rho_fg = rho_l - rho_g
            rho_m = (1 / rho_l + 1/(rho_g - rho_l)*chi)**(-1)

            # enthalpies
            hf0 = st.hL_p(p0)
            hg0 = st.hV_p(p0)
            hfg0 = hg0 - hf0

            # calculating p1
            pc1 = self.xi * self.ff * ics.g**2 / 2 / self.a_fluid / rho_m #topP1
            pc2 = rho_m * 9.81 # topP2
            pc3 = ics.g * qpp * self.xi / rho_fg / self.a_fluid / hfg0
            pc4 = chi * enthalpies.dhg(p0)
            pc5 = (1 - chi) * enthalpies.dhf(p0)
            pc6 = ics.g**2 / rho_fg / hfg0 * (pc4 + pc5)
#            pc6 = 1 - ics.g**2 / (rho_fg * enthalpies.hfg(p0)) * (chi * enthalpies.dhg(p0) + (1 - chi) * enthalpies.dhf(p0))

#            botp1 = chi * enthalpies.dhg(p0) + (1 - chi) * enthalpies.dhf(p0)
#            botp2 = 1 - ics.g**2 / (rho_fg * enthalpies.hfg(p0 * 1e-6)) * botp1
#            1 - g**2 / (rhofg * hfg) * (chi_e * dhgdp + (1 - chi_e) * dhfdp)

#            if i == 1: print(f"pc1: {pc1}\npc2: {pc2}\npc3: {pc3}\npc4: {pc4}\npc5: {pc5}\npc6: {pc6}")
#            if i == 1: print(chi_e0, enthalpies.dhg(p0), enthalpies.dhf(p0), ics.g, rho_fg, enthalpies.hfg(p0))
            
            dp = ((pc1 + pc2 + pc3) / (1 - pc6)) * dz * 1e-3
#            print(dp * 1e3, p0 * 1e6)
            p1 = p0 + dp
            
            # enthalpies 2
            hf1 = st.hL_p(p1)
            hg1 = st.hV_p(p1)
            hfg1 = hg1 - hf1

            dhf = hf1 - hf0
            dhg = hg1 - hg0
            dhfg = hfg1 - hfg0
            
            # calculating chi_e1
            chic1 = qpp * self.xi / self.a_fluid / ics.g / hfg0
            chic2 = chi * enthalpies.dhg(p1) * dp/dz
            chic3 = (1 - chi) * enthalpies.dhf(p1) * dp/dz
            chic4 = 1 / hfg0 * (chic2 + chic3)

            dchi_e = dz * (chic1 - chic4)
            chi_e1 = chi_e0 + dchi_e
            
            # calculating tf1
            t_sat = st.tsat_p(p1)
            
            tf1 = enthalpies.hfg(p1) * chi_e1 / self.cp + st.tsat_p(p1)

            # output
            if i % 10 == 0 and show: print(f"z: {round(z,5)} \ttf: {round(tf1,5)} \tp: {round(p1,5)} \tchi_e: {round(chi_e1,3)} \tt_sat: {round(t_sat,3)}")
            temp_fluid.append(tf1)
            pressure.append(p1)
            chi_e.append(chi_e1)
            temp_sat.append(t_sat)

        self.temp_fluid = temp_fluid
        self.temp_sat = temp_sat
        
        self.pressure = pressure
        self.chi_e = chi_e
        return

    def fluid_plotter(self):
        plt.plot(self.temp_fluid, self.zs)
        plt.plot(self.temp_sat, self.zs)
        plt.title("temps"), plt.grid('both')
        plt.show()

        plt.plot(self.pressure, self.zs)
        plt.title("pressure"), plt.grid('both')
        plt.show()
        plt.title("chi"), plt.grid('both')
        plt.plot(self.chi_e, self.zs)
        plt.show()

    def fluid_interpolaters(self, show):
        deg = 5
        
        self.func_temp_fluid = np.polynomial.Polynomial.fit(self.zs, self.temp_fluid, deg)
        self.func_temp_sat = np.polynomial.Polynomial.fit(self.zs, self.temp_sat, deg)
        self.func_pressure = np.polynomial.Polynomial.fit(self.zs, self.pressure, deg)
        self.func_chi_e = np.polynomial.Polynomial.fit(self.zs, self.chi_e, deg)

        if show:
            functions = [self.func_temp_fluid, self.func_temp_sat, self.func_pressure, self.func_chi_e]
            discrete = [self.temp_fluid, self.temp_sat, self.pressure, self.chi_e]
            titles = ["Fluid Temp [K]", "Saturation Temp [K]", "Pressure [MPa]", "Equilibrium Quality"]
        
            for func, disc, title in zip(functions, discrete, titles):
                plt.plot(func(self.zs), self.zs, label='Function', c='r')
                plt.plot(disc, self.zs, label = 'Discrete Points', ls=(0,(5,5)), c='b')
                plt.xlabel(title), plt.ylabel('Z position [m]')
                plt.legend()
                plt.show()

    def solve_clad(self):
        st = self.st
        ics = self.ics

        press = lambda z: self.func_pressure(z)
        F = lambda chi_e, p: 1 if chi_e < 0 else (1 + chi_e * self.pr * (st.rhoV_p(p) / st.rhoL_p(p) - 1))**(0.35)
        S = lambda chi_e, z: (1 + 0.055 * F(chi_e, press(z))**(0.1) * (ics.g * self.d_h / self.mu_f)**(0.16))**(-1)
        hnb = lambda z: 55 * (st.rhoV_p(press(z)) / st.rhoL_p(press(z)))**(0.12) * (1000 * self.heat_flux(z))**(2/3) * (-np.log(press(z) / self.pc)) * self.mw**(-0.5)
        
        def one_phase(z): return 1000 * self.heat_flux(z) / self.h_f0 + self.func_temp_fluid(z)
        
        def two_phase(tw, z, chi_e):
            f_ = (F(chi_e, press(z)) * self.h_f0 * (tw - self.func_temp_fluid(z)))**2
            s_ = (S(chi_e,z) * hnb(z) * (tw - st.tsat_p(press(z))))**2

#            print(f_, s_)
            
            return 1000 * self.heat_flux(z) - (f_ + s_)**(0.5)

        tcs = []
        guess = st.tsat_p(ics.p_z0)

        for i,z in enumerate(self.zs):
            chi_e = self.func_chi_e(z)
            
            t_guess = one_phase(z)
            tsat = self.func_temp_sat(z)
            
            if t_guess < tsat:
                tcs.append(t_guess)

            if t_guess >= tsat:
                try: _ = self.onb * 2
                except: self.onb = z

                func = lambda tw: two_phase(tw, z, chi_e)
                root = sci.optimize.root(func, guess).x[0]
                tcs.append(root)

#            print(tcs[i], st.tsat_p(self.func_pressure(z)))

        try: self.sat_boil = sci.optimize.root(self.func_chi_e, self.h_f0 / 3).x[0]
        except ValueError:
            pass
        self.func_tcs = sci.interpolate.interp1d(self.zs, tcs)

        #plt.plot(self.func_tcs(self.zs), self.zs)
        #plt.plot(self.func_temp_sat(self.zs), self.zs)
        #print(np.max(tcs))
        return

    def solve_fuel(self):
        ics = self.ics
        
        r1 = ics.d_fuel / 2
        r2 = r1 + ics.t_gap
        r3 = ics.d_rod / 2
        
        fuel_region = np.linspace(0,r1,1000)
        gap_region = np.linspace(r1,r2,1000)
        clad_region =  np.linspace(r3,r3,1000)
        
        c3 = lambda z: - 1e3 * self.vol_heat_gen(z) * r1**2 / 2 / ics.k_gap
        c5 = lambda z: ics.k_gap / ics.k_clad * c3(z)
        c6 = lambda z: self.func_temp_fluid(z) - c5(z) * np.log(r3)
        c4 = lambda z: c5(z) * np.log(r2) - c3(z) * np.log(r2) + c6(z)
        c2 = lambda z: 1e3 * self.vol_heat_gen(z) * r1**2 / 4 / ics.k_fuel + c3(z) * np.log(r1) + c4(z)

        t_fuel = lambda r,z: - 1e3 * self.vol_heat_gen(z) * r**2 / 4 / ics.k_fuel + c2(z)
        t_gap = lambda r,z: c3(z) * np.log(r) + c4(z)
        t_clad = lambda r,z: c5(z) * np.log(r) + c6(z)

        t_cl = t_fuel(0, self.zs)
        self.temp_cl = np.polynomial.Polynomial.fit(self.zs, t_cl, 10)
        return

    def plotter(self, save = False):
        plt.plot(self.temp_cl(self.zs), self.zs, label = "Centerline Temp")
        plt.plot(self.func_temp_sat(self.zs), self.zs, label = "Saturation Temp")
        plt.plot(self.func_tcs(self.zs), self.zs, label = "Clad Surface Temp")
        plt.plot(self.func_temp_fluid(self.zs), self.zs, label = "Fluid Temp")
        plt.grid("both"), plt.legend(), plt.ylabel("Z position [m]")
        plt.xlabel("Temperature [K]")
        if save: plt.savefig("plots/all-temps-vs-z.png", dpi=600)
        plt.show()

        plt.plot(self.func_temp_sat(self.zs), self.zs, label = "Saturation Temp")
        plt.plot(self.func_tcs(self.zs), self.zs, label = "Clad Surface Temp")
        plt.plot(self.func_temp_fluid(self.zs), self.zs, label = "Fluid Temp")
        plt.grid("both"), plt.legend(), plt.ylabel("Z position [m]")
        plt.xlabel("Temperature [K]")
        if save: plt.savefig("plots/temps-vs-z.png", dpi=600)
        plt.show()

        plt.plot(self.temp_cl(self.zs), self.zs, label = "Centerline Temp")
        plt.grid("both"), plt.legend(), plt.ylabel("Z position [m]")
        plt.xlabel("Temperature [K]")
        if save: plt.savefig("plots/tcl-vs-z.png", dpi=600)
        plt.show()

        plt.plot(self.func_chi_e(self.zs), self.zs, label = "Equilibrium Quality")
        plt.grid("both"), plt.legend(), plt.ylabel("Z position [m]")
        plt.xlabel("Equilibrium Quality")
        if save: plt.savefig("plots/chie-vs-z.png", dpi=600)
        plt.show()
        
        plt.plot(self.func_pressure(self.zs), self.zs, label = "Fluid Pressure")
        plt.grid("both"), plt.legend(), plt.ylabel("Z position [m]")
        plt.xlabel("Fluid Pressure [mPa]")
        if save: plt.savefig("plots/p-vs-z.png", dpi=600)
        plt.show()


a = Solver(1)
a.solve_fluid(0)
#a.fluid_plotter()
a.fluid_interpolaters(False)
a.solve_clad()
a.solve_fuel()
a.plotter(1)
```
