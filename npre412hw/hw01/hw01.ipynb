# Question 3

```python
# data
energy = np.array([43935276, 44557699, 46543126, 48409682, 50494169, 52577565,
                    55573431, 57335511, 60820699, 63791447, 65939114, 67255613,
                    70744691, 73788463, 71758674, 69810078, 73944199, 76351470,
                    77987575, 78844092, 76064555, 74195857, 70772226, 70452934,
                    74177235, 74340863, 74451441, 77112925, 81077213, 82703379,
                    82277974, 82212657, 83842847, 85220279, 87080163, 88732046,
                    91474194, 92105715, 92614839, 94212808, 96686060, 94391208,
                    95582476, 95806580, 98044608, 98108764, 97231396, 98987782,
                    96657655, 91632089, 95130010, 93972308, 91667414, 94237461,
                    95328233, 94472508, 94086508, 93906347, 97404148, 96576100,
                    88870847, 93349660, 94773722]) * 1e-6

years = np.arange(1960, 2022 + 1) - 1960


# projection function
m = (energy[-1] - energy[0]) / (len(energy) - 1)
b = energy[0]

proj_years = np.arange(1960, 2050 + 1) - 1960
proj_energy = m * proj_years + b

def get_energy(year):
    x = (year - 1960)
    return m * x + b


# calculating percent change
energy2025 = get_energy(2025)
energy2050 = get_energy(2050)

percent_change = (energy2050 - energy2025) / energy2025
print(f"Energy in 2025: {energy2025} Quads")
print(f"Energy in 2050: {energy2050} Quads")

print(f"Total % Change 2025-2050:  {round(percent_change * 100, 5)}%")
print(f"Annual % Change 2025-2050: {round(percent_change / 25 * 100, 5)}%")

# plotting
def plotter():
    plt.plot(years + 1960, energy, label="Data")
    plt.plot(proj_years + 1960, proj_energy, ls=(0,(5,3)), c='r', label="Projection")
    plt.grid('both')
    plt.ylabel('Total Energy Consumption [Quadrillion BTU]')
    plt.xlabel('Year')
    plt.legend()
    plt.savefig("hw01-q3.png", dpi=600)
    plt.show()

plotter()
```
