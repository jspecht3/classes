## Question 3

```python
q3d_raw = {}
for i in range(0, 101, 5):
    if i != 35 and i != 45 and i !=75 and i != 85 and i != 95:
        q3d_raw[i] = []


# getting all the data points
for file in glob.glob("data/inflections/*"):
    with open(file) as f:
        reader = csv.reader(f, delimiter = '\t')
        for row in reader:
            try:
                percent = int(row[0])
                to_append = []
                for element in row[1:]:
                    if element == '': break
                    if int(element) == 0: break
                    to_append.append(int(element))
                if to_append != []:
                    q3d_raw[percent].append(to_append)
            except: continue


# rejecting the ones that don't have the most inflections
for key in q3d_raw.keys():
    largest_array_size = 0
    for points in q3d_raw[key]:
        if len(points) > largest_array_size:
            largest_array_size = len(points)
    for i in reversed(range(len(q3d_raw[key]))):
        if len(q3d_raw[key][i]) != largest_array_size:
            q3d_raw[key].remove(q3d_raw[key][i])


# choosing the best data
q3d = {}

q3d[0] = q3d_raw[0][0]
q3d[5] = q3d_raw[5][2]
q3d[10] = q3d_raw[10][0]
q3d[15] = q3d_raw[15][1]
q3d[20] = q3d_raw[20][0]
q3d[25] = q3d_raw[25][3]
q3d[30] = q3d_raw[30][1]
q3d[40] = q3d_raw[40][0]
q3d[50] = q3d_raw[50][0]
q3d[55] = q3d_raw[55][0]
q3d[60] = q3d_raw[60][0]
q3d[65] = q3d_raw[65][1]
q3d[70] = q3d_raw[70][1]
q3d[80] = q3d_raw[80][1]
q3d[90] = q3d_raw[90][1]
q3d[100] = q3d_raw[100][1]
```

```python
q3_table = ''

for key in q3d.keys():
    to_append = str(key)
    for i in range(3):
        try: to_append += ' & ' + (str(q3d[key][i]))
        except: to_append += ' & -'
    q3_table += to_append + " \\\ \hline "
    print(to_append)

#print(q3_table)
```
