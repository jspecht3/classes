```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import csv
```

```python
data = {}

skips = ['System ID', 'Method name only', 'Test type','Sample file name','Last test date','Last test time',
         'Course','Lab Section Number','Other Information','Crosshead Speed','Unwelded Gap Thickness tg (mm)',
         'Notes During Test','Results Table','Weld Type','Results Table 1','Material','Geometry','','Time','(s)']

for file in glob("data/*"):
    name = file.split('data/N02')[1].split('_1.csv')[0]
    if (name[0] == '-'): name = name[3:]
    if (name[0] == 'B'): name = name[1:]

    mat = name[:4]
    weld = (list(filter(None, name[4:].split('-')))[0]).lower()

    if mat not in data: data[mat] = {}
    if weld not in data[mat]: data[mat][weld] = {}
    
    with open(file) as f:
        reader = csv.reader(f)
        for row in reader:
            if row == []: continue
            if row[0] in skips: continue

            if row[0] == 'Group':
                group = row[-1]
                data[mat][weld][group] = {}
                
                data[mat][weld][group]['time'] = []
                data[mat][weld][group]['disp'] = []
                data[mat][weld][group]['force'] = []
                continue

            if row[0] == 'Width of Specimen and Weld w':
                data[mat][weld][group]['w'] = float(row[-1])
                continue

            if row[0] == 'Specimen Thickness t':
                data[mat][weld][group]['t'] = float(row[-1])
                continue

            if row[0] == 'Weld Thickness tw':
                data[mat][weld][group]['tw'] = float(row[-1])
                continue

            if row[0] == 'Specimen Gage Length':
                data[mat][weld][group]['l'] = float(row[-1])
                continue

            if row[0] == 'Fracture Type and Location':
                try:
                    data[mat][weld][group]['t-gap'] = float(row[-1].split(' ')[-2])
                    continue
                except: continue

            if row[0] == 'Maximum Force':
                data[mat][weld][group]['fmax'] = float(row[-1])
                continue

            # writing to lists
            data[mat][weld][group]['time'].append(float(row[0]))
            data[mat][weld][group]['disp'].append(float(row[1]))
            data[mat][weld][group]['force'].append(float(row[2]))
```

```python
for mat in data:
    for group in data[mat]['cham']:
        if group == 'A': continue

        force = data[mat]['cham'][group]['force']
        disp = data[mat]['cham'][group]['disp']
        
        t = data[mat]['cham'][group]['t'] / 1000
        w = data[mat]['cham'][group]['w'] / 1000
        
        area = t * w
        stress = np.array(force) / area
        
        l = data[mat]['cham'][group]['l']
        strain = np.array(disp) / l


        if mat == '6061': label = "6061 Aluminum Chamfer"
        if mat == '1018': label = "1018 Steel Chamfer"
            
        plt.plot(strain, stress / 1e3, label = label)

plt.xlabel('Strain')
plt.ylabel('Stress [MPa]')
plt.grid('both')
plt.legend()
plt.savefig('plots/q2.png', dpi=600)
plt.show()
```

```python
def q2_props(mat):
    force = data[mat]['cham']['C']['force']
    disp = data[mat]['cham']['C']['disp']
    
    t = data[mat]['cham']['C']['t'] / 1000
    w = data[mat]['cham']['C']['w'] / 1000
    
    area = t * w
    stress = np.array(force) / area
    
    l = data[mat]['cham'][group]['l']
    strain = np.array(disp) / l

    return stress / 1000, strain
```

```python
q2d = {}

for file in glob('other/*.csv'):
    mat = (file.split('other/')[1].split('_')[0])
    q2d[mat] = {'stress': [], 'strain': []}

    go = False

    for row in csv.reader(open(file)):
        if row == []: continue
        if row[0] == 'Area':
            area = float(row[1])
            continue
        
        if go:
            stress = float(row[-2]) / area
            strain = float(row[-1])

            if stress > 0:
                q2d[mat]['stress'].append(stress)
                q2d[mat]['strain'].append(strain)

        if row[0] == '(sec)':
            go = True
            continue
```

```python
for mat in q2d:
    stress, strain = q2d[mat]['stress'], q2d[mat]['strain']

    if mat == '6061': label = "6061 Aluminum Chamfer"
    if mat == '1018CR': label = "1018 Steel Chamfer"
    
    plt.plot(strain,stress, label=label)

plt.grid('both')
plt.xlabel('Strain')
plt.ylabel('Stress [MPa]')
plt.legend()
plt.savefig('plots/q2-raw.png', dpi=600)
plt.show()
```

```python
import scipy.interpolate as interp

q4d = {'1018CR': {}, '6061': {}}

for file in glob('other/*vickers*'):
    mat = file.split('other/')[1].split('_vickers')[0]
    
    with open(file) as f:
        go = False
        
        xs = []
        ys = []

        for row in csv.reader(f, delimiter='\t'):
            if row == []: continue

            if go:
                xs.append(float(row[1]))
                ys.append(float(row[2]))
            
            if row[0] == 'Number' or row[0] == 'Distance(mm)':
                go = True

        q4d[mat]['x'] = xs
        q4d[mat]['y'] = ys
        
        q4d[mat]['d'] = []
        for x,y in zip(xs,ys):
            q4d[mat]['d'].append((x + y) / 2)

for mat in q4d:
    vhn = 1.85 * 500 / np.array(q4d[mat]['d'])**2 * 1000
    xs = np.arange(len(vhn))

    spline = interp.make_smoothing_spline(xs, vhn, lam=2)
    ys = spline(xs)

    if mat == '1018CR':
        plt.axvline(16, color='r', ls=(0,(5,3)), label = 'Heat Affected Zone')
        plt.axvline(21, color='r', ls=(0,(5,3)))

    if mat == '6061':
        plt.axvline(17, color='r', ls=(0,(5,3)), label = 'Heat Affected Zone')
        plt.axvline(23, color='r', ls=(0,(5,3)))
    
    plt.scatter(xs, vhn, label="Vickers Hardness")
    plt.plot(ys, c='k', label="Curve Fit")
    plt.xlabel('Number')
    plt.ylabel('Vickers Hardness Number')
    plt.legend()
    plt.savefig(f'plots/q4-{mat}.png', dpi=600)
    plt.show()
```
