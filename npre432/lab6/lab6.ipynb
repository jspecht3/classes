```python
import numpy as np
import csv
from glob import glob
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

# Data Processing

```python
files = glob("data/*")
data = {}

for file in files[:]:
    splitter = (file.split('/')[1].split("N02-")[1].split('.csv')[0].split('-'))
    splitter.remove(splitter[0])

    # getting temp treatment
    temp = splitter[-1]
    splitter.remove(splitter[-1])

    # getting material type
    material = ""
    for m in splitter:
        material += m + '-'
    material = material[:-1]

    # dict index
    if material not in data.keys(): 
        data[material] = {}
    if temp not in data[material].keys(): 
        data[material][temp] = {1: {}, 2: {}}
    
    index = 1
    if data[material][temp][1] != {}:
        index = 2

    with open(file, mode = 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if row == []: continue
            
            if row[0] == '[Shared Parameters]': continue
            if row[0] == 'Comment': continue
            if row[0] == '[Shared Data]': continue
            if row[0] == '[Step [1] Data]': continue
            if row[0] == 'Date': continue
            if row[0] == 'Instrument': continue
            if row[0] == 'Start Time': continue
            if row[0] == 'Stop Time': continue
            if row[0] == 'Test Name': continue
            if row[0] == 'Specimen ID': continue
            
            data[material][temp][index][row[0]] = row[1]
```

# Question 1

- plot the impact energy as a function of temperature
- include all materials
- plot max load 

## data

```python
mats = ['1045HR', '2024-T4', 'PMMA', 'HDPE']
temps = ['0C', 'RT', 'BW']

q1data = {}

for mat in mats:
    if mat not in q1data.keys(): q1data[mat] = {'energy': {}, 'load': {}}

    for index in [1,2]:
        if index not in q1data[mat]['energy'].keys():
            q1data[mat]['energy'][index] = []
            q1data[mat]['load'][index] = []
            
        energies = []
        loads = []
        
        for temp in temps:
            if data[mat][temp][index] == {}: continue
            energies.append(float(data[mat][temp][index]['Total Energy']))
            loads.append(float(data[mat][temp][index]['Peak Force']) / 1000)

        q1data[mat]['energy'][index] = energies
        q1data[mat]['load'][index] = loads
```

## plotting

```python
xs = [0, 22, 100]

for glip in ['energy', 'load']:
    for index in [1,2]:
        for mat in mats:
            q1ys = q1data[mat][glip][index]

            if mat == '1045HR': label = '1045 Hot-Rolled Steel'
            if mat == '2024-T4': label = '2024-T4 Aluminum'
            if mat == 'PMMA': label = 'Polymethyl Methacrylate'
            if mat == 'HDPE': label = 'High Density Polyethylene'

            if len(xs) == len(q1ys): plt.plot(xs, q1ys, label=label)
            if len(xs) != len(q1ys): plt.plot(xs[:2], q1ys, label=label)

        plt.xlabel('Temperature [$\degree$C]')
    
        if glip == 'energy':
            plt.ylabel('Absorbed Impact Energy [J]')
            if index == 1: plt.legend(loc=[0.01,0.37])
            if index == 2: plt.legend(loc=[0.01,0.5])
        if glip == 'load':
            plt.ylabel('Peak Force [kN]')
            plt.legend(loc=[0.52,0.75])
            plt.ylim([0,20])
        
        plt.grid('both')
        plt.savefig(f"plots/q1{glip}{index}.png", dpi=600)
        plt.show()
```

```python
q2data = {}

for mat in mats:
    q2data[mat] = {}
    for temp in temps:
        q2data[mat][temp] = {}
        
        energy = []
        load = []
        time = []
        
        for index in [1,2]:
            if data[mat][temp][index] == {}: continue

            energy.append(float(data[mat][temp][index]['Total Energy']))
            load.append(float(data[mat][temp][index]['Peak Force']))
            time.append(float(data[mat][temp][index]['Total Time']))

        q2data[mat][temp]['energy'] = np.mean(energy)
        q2data[mat][temp]['load'] = np.mean(load)
        q2data[mat][temp]['time'] = np.mean(time)
```

# Question 4

- plot the load-displacement curve for both mats
- determine the type of curve you have

```python
files = glob('data/kic/*_d.txt')
q4data = {}

for file in files:
    mat = file.split('data/kic/N02')[1].split('_d.txt')[0][2:]
    q4data[mat] = {
        'position': [],
        'load': [],
        'disp': []
    }
    
    with open(file, 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            if row == []: continue
            if row[0] == '*': continue
            if row[0] == 'POSN mm \tLOAD kn\tCTOD mm': continue

            row = row[0].split('\t')
            
            q4data[mat]['position'].append(float(row[0]))
            q4data[mat]['load'].append(float(row[1]))
            q4data[mat]['disp'].append(float(row[2]))
```

- w = width
- a = w - \bar{z}

```python
# 2024
def plotter2024():
    disp = q4data['2024']['disp']
    load = q4data['2024']['load']
    
    plt.plot(disp, load, label='2024 Al LD Curve')
    
    index = 250
#    plt.axvline(disp[index], ls=(0,(5,3)), c='k', lw=1.5)

    m = load[index] / disp[index] * 0.95
    xs = np.linspace(0,0.75, 1000)
    ys = m * xs

    plt.plot(xs, ys, c='r', label="95% Secant Line")
#    plt.xlim([0.5,0.6])
#    plt.ylim([20,25])

    pq = load[294]
    plt.axvline(disp[294], ls=(0,(5,3)), c='k', lw=1.5, label=f'P$_Q$: {round(pq,5)}')

    pmax = disp[np.argmax(load)]
    plt.axvline(pmax, ls=(0,(5,3)),
                c='g', lw=1.5,
                label='P$_{max}$: ' + str(round(np.max(load),5)))

    q4data['2024']['pq'] = pq
    q4data['2024']['pmax'] = np.max(load)
    q4data['2024']['type'] = 1
    
    print(f"Type 1\nP_q: {pq}\nP_max: {np.max(load)}")
    plt.xlabel('Displacement [mm]')
    plt.ylabel('Load [kN]')
    plt.grid('both')
    plt.legend()
    plt.savefig('plots/q42024.png', dpi=600)
    plt.show()

plotter2024()
```

```python
# 7075
def plotter7075():
    disp = q4data['7075']['disp']
    load = q4data['7075']['load']

    lmax = disp[np.argmax(load)]
    plt.axvline(lmax, ls=(0,(5,3)), c='k', lw=1.5,
                label=f'P$_Q$ and ' + 'P$_{max}$ : ' + str(round(np.max(load),5)))
    plt.plot(disp, load, label='7075 Al LD Curve')

    q4data['7075']['pq'] = np.max(load)
    q4data['7075']['pmax'] = np.max(load)
    q4data['7075']['type'] = 3
    
    print(f"Type 3\nP_q = P_max: {np.max(load)}")
    plt.xlabel('Displacement [mm]')
    plt.ylabel('Load [kN]')
    plt.grid('both')
    plt.legend()
    plt.savefig('plots/q47075.png', dpi=600)
    plt.show()

plotter7075()
```

```python
q4data['2024']['w'] = 50.90 / 1000
q4data['2024']['a'] = (50.90 - np.mean([26.66, 26.43, 26.93, 27.47, 26.69, 27.22])) / 1000
q4data['2024']['pp'] = 10
q4data['2024']['t'] = 25.39 / 1000

q4data['7075']['w'] = 51.05 / 1000
q4data['7075']['a'] = (51.05 - np.mean([26.52, 25.98, 26.92, 27.78, 26.07, 27.44])) / 1000
q4data['7075']['pp'] = 8
q4data['7075']['t'] = 25.27 / 1000
```

```python
def f_aw(a,w):
    aw = a / w
    print(f"aw {aw}")
    
    ret1 = (2 + aw) / ((1 - aw)**3)**(1/2)
    ret2 = (0.886 + 4.644*aw - 13.32*aw**2 + 14.72*aw**3 - 5.6*aw**4)
    
    return ret1 * ret2
```

```python
for mat in q4data:
    print(mat, q4data[mat]['a'] * 1000)
```

```python
def c1(mat):
    pp = q4data[mat]['pp']
    pq = q4data[mat]['pq']

    boop = pp < 0.6 * pq

    strout = f"P_precrack is {boop * '<'}{(not boop) * '>'} 0.6P_q, {boop}"
    print(strout, pp, 0.6 * pq)
    return

def c2(mat):
    pmax = q4data[mat]['pmax']
    pq = q4data[mat]['pq']

    boop = pmax < 1.1 * pq

    strout = f"P_max is {boop * '<'}{(not boop) * '>'} 1.1P_q, {boop}"
    print(strout, pmax, 1.1*pq)
    return

def kq(mat):
    pq = q4data[mat]['pq']
    t = q4data[mat]['t']
    w = q4data[mat]['w']
    a = q4data[mat]['a']

    return pq / t / w**(1/2) * f_aw(a,w)

def c3(mat):
    kq3 = kq(mat)
    t = q4data[mat]['t']
    w = q4data[mat]['w']
    a = q4data[mat]['a']
    pmax = q4data[mat]['pmax']

    area = np.pi * a**2
    compare = 2.5 * (kq3 * area / pmax)**2 / 1000

    boop1 = a > compare
    boop2 = (w - a) > compare
    boop3 = t > compare

    print(f" 3 {boop1}, {boop2}, {boop3}", a * 1000, (w-a) * 1000, t * 1000, compare * 1000)
    return

for mat in ['2024', '7075']:
    print(f"----- {mat} -----")
    c1(mat)
    c2(mat)
    c3(mat)
    print(f"Kq = {kq(mat) / 1000}")
    print("\n")
```
